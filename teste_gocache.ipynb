#### 1. Análise Exploratório dos Dados

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Obter o diretório atual do notebook
diretorio_atual = os.getcwd()

# Concatenar o caminho para a pasta "dados"
caminho_pasta_dados = os.path.join(diretorio_atual, 'dados')
```

##### Estatísticas Descritivas

Vamos analisar estatísticas descritivas dos dados para entender melhor sua distribuição e identificar possíveis padrões anômalos.

```python
# Carregar o arquivo sem ataque
dados_sem_ataque = pd.read_csv(caminho_pasta_dados + '\\arquivo_sem_ataque.csv')
dados_com_ataque = pd.read_csv(caminho_pasta_dados + '\\arquivo_com_ataque.csv')

# Exibir as primeiras linhas de cada dataframe para entender a estrutura dos dados
print("Dados sem ataque:")
display(dados_sem_ataque.head())

print("\nDados com ataque:")
display(dados_com_ataque.head())
```

```python
# Estatísticas descritivas dos dados sem ataque
print("Estatísticas descritivas dos dados sem ataque:")
print(dados_sem_ataque.describe())

# Estatísticas descritivas dos dados com ataque
print("Estatísticas descritivas dos dados com ataque:")
print(dados_com_ataque.describe())
```

##### Visualização dos Dados

Para identificar visualmente o aumento repentino de tráfego, vamos criar gráficos de linha para a quantidade de requisições ao longo do tempo.

```python
# Converter a coluna 'hora' para datetime
dados_sem_ataque['hora'] = pd.to_datetime(dados_sem_ataque['hora'])
dados_com_ataque['hora'] = pd.to_datetime(dados_com_ataque['hora'])

# Contar o número de requisições a cada 5 segundos
dados_sem_ataque['contagem'] = 1
dados_com_ataque['contagem'] = 1

# Resumir os dados por intervalo de 5 segundos
dados_sem_ataque_resumido = dados_sem_ataque.set_index('hora').resample('5S').sum()
dados_com_ataque_resumido = dados_com_ataque.set_index('hora').resample('5S').sum()

# Plotar os gráficos
plt.figure(figsize=(15, 6))
plt.plot(dados_sem_ataque_resumido.index, dados_sem_ataque_resumido['contagem'], label='Sem ataque')
plt.plot(dados_com_ataque_resumido.index, dados_com_ataque_resumido['contagem'], label='Com ataque')
plt.xlabel('Hora')
plt.ylabel('Número de Requisições')
plt.title('Número de Requisições ao Longo do Tempo')
plt.legend()
plt.show()
```

```python
# Plotar os gráficos
plt.figure(figsize=(15, 6))
plt.plot(dados_sem_ataque_resumido.index, dados_sem_ataque_resumido['tamanho_request'], label='Sem ataque')
plt.plot(dados_com_ataque_resumido.index, dados_com_ataque_resumido['tamanho_request'], label='Com ataque')
plt.xlabel('Hora')
plt.ylabel('Tamanho das Requisições')
plt.title('Tamanho das Requisições ao Longo do Tempo')
plt.legend()
plt.show()
```

```python
# Plotar os gráficos
plt.figure(figsize=(15, 6))
plt.hist(dados_sem_ataque_resumido['tamanho_request'], label='Sem ataque')
plt.title('Distribuição dos dados de acordo com as requisições')
plt.legend()
plt.show()
```

```python
# Plotar os gráficos
plt.figure(figsize=(15, 6))
plt.hist(dados_com_ataque_resumido['tamanho_request'], label='Com ataque')
plt.title('Distribuição dos dados de acordo com as requisições')
plt.legend()
plt.show()
```

#### 2. Identificação de Padrões de Ataque

##### Análise de Características

Vamos explorar diferentes características dos logs (como IP, user agent, URI, etc.) para identificar padrões comuns durante o ataque.

```python
# Análise de IPs únicos
ips_sem_ataque = dados_sem_ataque['ip'].nunique()
ips_com_ataque = dados_com_ataque['ip'].nunique()

print(f"Número de IPs únicos sem ataque: {ips_sem_ataque}")
print(f"Número de IPs únicos com ataque: {ips_com_ataque}")

# Análise de user agents únicos
user_agents_sem_ataque = dados_sem_ataque['useragent'].nunique()
user_agents_com_ataque = dados_com_ataque['useragent'].nunique()

print(f"Número de user agents únicos sem ataque: {user_agents_sem_ataque}")
print(f"Número de user agents únicos com ataque: {user_agents_com_ataque}")
```

#### 3. Preparação dos Dados

Para desenvolver um modelo de classificação, precisamos preparar os dados. Vamos codificar as variáveis categóricas e dividir os dados em conjuntos de treinamento e teste.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Seleção de atributos
atributos = ['ip', 'metodo', 'protocolo', 'hostname', 'uri', 'querystring', 'status_code', 'useragent', 'organizacao', 'pais', 'cidade']
X = dados_com_ataque[atributos]

# Codificação das variáveis categóricas
label_encoders = {}
for column in X.columns:
    label_encoders[column] = LabelEncoder()
    X[column] = label_encoders[column].fit_transform(X[column])

# Criação da variável alvo (1 para ataque, 0 para tráfego normal), pensando numa quantidade média sem ataque, o que estiver acima da média é suspeito de ataque.
y = (dados_com_ataque['tamanho_request'] >= dados_sem_ataque['tamanho_request'].mean()).astype(int)
```

```python
y.value_counts()
```

```python
# Divisão dos dados em treinamento e teste de forma estratificada
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
print(X_train.shape,  y_train.shape)
print(X_test.shape,  y_test.shape)
```

#### 4. Treinamento do Modelo

Vamos treinar um modelo de classificação usando uma árvore de decisão.

```python
from sklearn.metrics import accuracy_score, f1_score, roc_curve, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Treinamento do modelo baseado em Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# Avaliação do modelo
y_pred = rf.predict(X_test)

# Obter as probabilidades das classes
y_prob = rf.predict_proba(X_test)[:, 1]

print(f"Acurácia......: {round(accuracy_score(y_test, y_pred), 4)*100}")
print(f"F1 score......: {round(f1_score(y_test, y_pred), 4)*100}")
print()
print("Resultados por classes")
print()
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

# Calcular a curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

# Plotar a curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
